# Temperature Data

```python
# https://www.kaggle.com/berkeleyearth/climate-change-earth-surface-temperature-data?select=GlobalLandTemperaturesByState.csv
temp_df = pd.read_csv('GlobalLandTemperaturesByState.csv')
```

```python
temp_df = temp_df[temp_df['Country'] == 'United States'] # only dealing with wildfires in the US
temp_df.drop(['Country'], axis=1, inplace=True)
```

```python
def trimYears(x):
    year = int(x.split('-')[0])
    if year > 2015 or year < 1992:
        return 'bye'
    return x

def getYear(x):
    return int(x.split('-')[0])  

def getMonth(x):
    return int(x.split('-')[1])  
```

```python
# trim the years to be between 1992 and 2015 (this dataset only goes through 2013)
# not necessary due to how the merge is done later...but why not do it anyways
print(len(temp_df))
temp_df['dt'] = temp_df['dt'].apply(trimYears)
temp_df = temp_df[temp_df['dt'] != 'bye']
print(len(temp_df))
temp_df.head()
```

```python
temp_df.loc[:, 'Year'] = temp_df['dt'].apply(getYear)
temp_df.loc[:, 'Month'] = temp_df['dt'].apply(getMonth)
temp_df.head()
```

```python
# Drop 'dt' and replace state with abbreviation
temp_df.drop(['dt'], axis=1, inplace=True)

temp_df = temp_df[ temp_df['State'] != 'District Of Columbia']
temp_df['State'] = temp_df['State'].apply(lambda x: x if x != 'Georgia (State)' else 'Georgia')
temp_df['State'] = temp_df['State'].apply(lambda x: us.states.mapping('name', 'abbr')[x])
print(len(temp_df['State'].unique()))
temp_df.head()
```

# Join

```python
# Need to get month from year and date of year (1-366) to join
ffires_df['Holder'] = ffires_df.T.apply(lambda x: str(x['DISCOVERY_DOY']) + " " + str(x['FIRE_YEAR']))
ffires_df['Month'] = ffires_df['Holder'].apply(lambda x: datetime.datetime.strptime(x, '%j %Y').month)
ffires_df.drop(['Holder', 'DISCOVERY_DOY'], axis=1, inplace=True)
ffires_df.head()
```

```python
# Join them
print(len(ffires_df))
joined_df = ffires_df.merge(temp_df, left_on=['FIRE_YEAR', 'Month', 'STATE'], right_on=['Year', 'Month', 'State'], how='left')
print(len(joined_df))
joined_df.head()
```

```python
# Drop extra columns
joined_df.drop(['Year', 'State'], axis=1, inplace=True)
joined_df.head()
```
